# Binary thresholding method on histogram of pixel intensity
   image :-  two region of inerest ,object and background. So ,in this case ,histogram of pixel intensity would 
              represent a probability distribution that is bimodal ,i.e have high density around two pixel
              intensity values. 
# Idea:
                i) choose a threshold pixel value and set all pixel value below the threshold as 255 and above
                the threshold as 0.
                ii) this idea would ensure that we have object and background , representes by white and black
                color.

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('image1.jpeg')
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img ,cmap ='gray')
```

```python
row ,col = np.shape(gray_img)         # storing the shape of image
gray_flat = np.reshape(gray_img ,(row*col ,1))[: ,0]
print(np.shape(gray_flat))
```

```python
ax = plt.subplot(222)
ax.hist(gray_flat ,color ='gray')
```

As we can see the density  is minimum around 100. so, choosing threshold intensity ,t= 100

```python
t= 100
gray_count =[]
for i in range(len(gray_flat)):
    if gray_flat[i] <t:
        gray_count.append(255)
    else:
        gray_count.append(0)
gray_img1 = np.reshape(np.array(gray_count) ,(row ,col))
bx = plt.subplot(333)
bx.imshow(gray_img1 ,cmap ='gray')
```

# OTSU's Method(for gray-scale/one color channel image)
Idea: Determine the threshold by maximizing the variance between different segment of the images

# Step
    i)  compute the probability of each pixel intensity in the image.
                p(i) = count(i)/n   , where i =intensity ->[0,255] , n = total no of pixel
    ii) if the image has two segment c1 and c2 based on threshold t ,then the set of pixels {0 ,1,2...t}
        belong to c1 while set of pixels {t+1 ,t+2,...,L-1} belongs to c2. The variance between the two segment 
        is determined by the sum of square deviation of the mean of hte cluster w.r.t cluster.
                            var(c1 ,c2) = P(c1)*(u1 -u)**2 + P(c2)*(u2 -u)**2
                     where,   P(c1) = sigma_0_to_t(P(i)) ,P(c2) = sigma_t+1_to_L-1(P(i))
                              u1 = sigma_0_to_t(P(i)*i)  , u2 = sigma_t+1_to_L-1(P(i) *i)
                              u = sigma_0_to_L-1(P(i)*i)
     iii) P(c1) ,P(c2) ,u1 ,u2 is function of t. t_bar = Argmax var(c1,c2).
          instead of computing derivative and the setting it to zero to obtain t_bar ,we can evaluate 
          var(c1 ,c2) for all values of t {0 ,1 ,.....,255)  and choose t_bar at which var(c1 ,c2) is maximum.
       

```python
img = cv2.imread('3.jpeg')
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img ,cmap ='gray')
```

```python
row ,col = np.shape(gray_img) # storing the shape of image
hist_dist = 256*[0]# for stroing the probability of each pixels
for i in range(row):
    for j in range(col):
        hist_dist[gray_img[i ,j]] += 1 # counting the no of pixels
hist_dist = [c/float(row*col) for c in hist_dist]    # finding the probability
plt.plot(hist_dist)
```

```python
print(row ,col)
def Var_c1_c2(hist_dist ,t):
    u1 ,u2,u ,p1 ,p2 = 0 ,0 ,0 ,0 ,0
    for i in range(t+1):
        u1 += hist_dist[i]*i
        p1 += hist_dist[i]
    for i in range(t+1 ,256):
        u2 += hist_dist[i]*i
        p2 +=  hist_dist[i]
    for i in range(256):
        u += hist_dist[i]*i
    var_c1_c2 = p1*(u1-u)**2 + p2*(u2 -u)**2
    return var_c1_c2
variance_list =[]                       # computing the varince for all values of t
for i in range(256):
    var_c1_c2 = Var_c1_c2(hist_dist ,i)
    variance_list.append(var_c1_c2)
t_bar = np.argmax(variance_list)              # choosing the value of t which has maximum variance
print(t_bar)

gray_img1 = np.zeros((row ,col))
print(np.shape(gray_img1))
for i in range(row):
    for j in range(col):
        if gray_img[i ,j] <= t_bar:
            gray_img1[i ,j] = 255
        else:
            gray_img1[i ,j] = 0
plt.imshow(gray_img1 ,cmap ='gray')
    
```

# Image segmentation using K-means clustering
    Idea :- i) K, a parameter that determine the number of different cluster to be formed.
            ii) Each cluster is represented by its cluster centroids based on specific input features.
                    input -> u == [width(intensity) ,height(intensity) ,channel(intensity)] 
                   
            

```python
# python implementation
import numpy as np
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('3.jpeg')
img_gray1 = cv2.cvtColor(img ,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray1 ,cmap ='gray')
print(img.shape)
```

```python
row ,col ,depth = img.shape
# collapsing the img into row*col ,depth for faster matrix operation
img_new = np.zeros((row*col ,3))
glob_ind =0
for i in range(row):
    for j in range(col):
        x = np.array([img[i ,j,0] ,img[i,j,1] ,img[i,j,2]])
        img_new[glob_ind,:] = x
        glob_ind += 1
####################
K = 10                                      # setting the no of clusters
num_iter =20 
for g in range(num_iter):
    # defining the cluster for stroing the cluster number and out_dist to store the distance from centroids
    clusters = np.zeros((row*col ,1))
    out_dist = np.zeros((row*col ,K))
    centroids = np.random.randint(0 ,255,size =(K ,3))
    for k in range(K):
        diff = img_new - centroids[k ,:]       # calculating the distance of a  particular centroid from each pixel(return shape - row*col,3)
        diff_dist = np.linalg.norm(diff ,axis =1)  # normalizing the difference (return shape - row*col ,1)
        out_dist[: ,k] = diff_dist             # storing the distance from a particular centroid
    # assigning the cluster with minimum distance from pixel loaction
    clusters = np.argmin(out_dist ,axis =1)
    # recomputing the centroids
    for k1 in np.unique(clusters):
        centroids[k1 ,:] = np.sum(img_new[clusters==k1 ,:] ,axis =0)/np.sum([clusters==k1])
    # rehspaing the clusters int 2-dim img for'
    clusters = np.reshape(clusters ,(row ,col))
    out_img = np.zeros(img.shape)
    # forming the 3-D image with the pixel replaced by theri centroids pixel intensity
    for i in range(row):
        for j in range(col):
            out_img[i ,j ,0] = centroids[clusters[i ,j] ,0]
            out_img[i ,j ,1] = centroids[clusters[i ,j] ,1]
            out_img[i ,j ,2] = centroids[clusters[i ,j] ,2]
            
out_img = np.array(out_img, dtype='uint8')
# displaying the output images
plt.imshow(out_img)
    
```
